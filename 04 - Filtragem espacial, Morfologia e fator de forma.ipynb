# Filtros no dominio espacial

http://www.scipy-lectures.org/advanced/image_processing/#image-filtering

```python
import numpy as np
from scipy import misc
from scipy import ndimage

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('figures/cameraman.tif')

imgplot = plt.imshow(img, 'gray')
plt.show()
```

```python
img=mpimg.imread('figures/Lenna.png')

img = (np.sum(img, axis=2)) / 3
imgplot = plt.imshow(img, 'gray')
plt.show()
```

### Filtro Gaussiano

```python
blurred_img = ndimage.gaussian_filter(img, sigma=2)

imgplot = plt.imshow(blurred_img, 'gray')
plt.show()
```

```python
blurred_img = ndimage.gaussian_filter(img, sigma=5)

imgplot = plt.imshow(blurred_img, 'gray')
plt.show()
```

## Filtro Uniforme

```python
blurred_img = ndimage.uniform_filter(img, size=10)

imgplot = plt.imshow(blurred_img, 'gray')
plt.show()
```

## Sharpening

```python
borders = img - blurred_img

# imgplot = plt.imshow(borders, 'gray')
imgplot = plt.imshow(abs(borders), 'gray')
# imgplot = plt.imshow(borders, 'gray', vmin=0, vmax=255)
plt.show()
```

```python
shaped = img + abs(borders) * 2

imgplot = plt.imshow(shaped, 'gray')
plt.show()
```

# Morfologia

```python
a = np.zeros((7,7), dtype=np.int)
a[1:6, 2:5] = 1
print(a)
imgplot = plt.imshow(a, 'gray')
plt.show()
```

```python
el = structure=np.ones((3,3))
el
```

### Erosion

```python
eroded = ndimage.binary_erosion(a, el).astype(a.dtype)
print(eroded)
imgplot = plt.imshow(eroded, 'gray')
plt.show()
```

### Dilation

```python
dilate = ndimage.binary_dilation(a, el).astype(a.dtype)
print(dilate)
imgplot = plt.imshow(dilate, 'gray')
plt.show()
```

## Close = Dilate + Erode

```python
a = np.zeros((15,15), dtype=np.int)
a[2:-2, 2:-2] = 1
a[4:6, 3:-5] = 0
print(a)
imgplot = plt.imshow(a, 'gray')
plt.show()
```

```python
dilate = ndimage.binary_dilation(a, el).astype(a.dtype)
eroded = ndimage.binary_erosion(dilate, el).astype(a.dtype)

print(eroded)
imgplot = plt.imshow(eroded, 'gray')
plt.show()
```

## Open = Erode + Dilate

```python
eroded = ndimage.binary_erosion(a, el).astype(a.dtype)
dilate = ndimage.binary_dilation(eroded, el).astype(a.dtype)


print(eroded)
imgplot = plt.imshow(eroded, 'gray')
plt.show()
```

# Fator de forma

### Perimetro^2 / area


```python
size = 200
center = size / 2
radius = 95

r,c = np.meshgrid(np.arange(0,size), np.arange(0,size), indexing='ij')
circle = ((r - center)**2 + (c - center)**2)**(1/2) < radius

imgplot = plt.imshow(circle, 'gray')
plt.show()
```

```python
eroded = ndimage.binary_erosion(circle, np.ones((3,3))).astype(a.dtype)
imgplot = plt.imshow(abs(circle - eroded), 'gray')
plt.show()
```

```python
area = np.sum(circle)
print('area = ', area)

perimetro = np.sum(abs(circle - eroded))
print('perimetro = ', perimetro)
(perimetro**2)/area
```
