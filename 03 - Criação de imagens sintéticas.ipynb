# Criação de imagens sintéticas

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
```

```python
r,c = np.meshgrid(np.arange(-75,75), np.arange(-100,100), indexing='ij')
f = r * c
f.max()
```

```python
imgplot = plt.imshow(f, 'gray')
plt.contour(c+100,r+75,f,20)
plt.show()
```

```python
fig = plt.figure(figsize=(5,5))
r,c = np.indices( (200,200))
f = r*np.sin(c/15)
imgplot = plt.imshow(f, 'gray')
plt.contour(c,r,f,5)
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')

ax.contour(c,r,f,5)
surf = ax.plot_surface(c, r, f, cmap=cm.coolwarm, linewidth=0, antialiased=True, alpha=0.3)


plt.show()
```
